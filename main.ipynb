## Add a column (Get teleop points scored)
This one may seem a little bit complicated, but I challenge you to look closer. The way this works to make the new column is:
df.assign(name=column).
in this case, name is "points_earned"
and the column is defined by the sum of all the ways to score points times their point values!

```python
df = df.assign(points_earned=
    (df.get("inner port")*3)+
    (df.get("outer port")*2)+
    (df.get("bottom port")*1)+
    (df.get("auton inner port")*6)+
    (df.get("auton outer port")*4)+
    (df.get("auton bottom port")*2)+
    (df.get("hang under shield")*25)+
    (df.get("auton motion")*5)+
    (df.get("park under shield")*5))
df.get("points_earned")
```

## Defense Score 
This score is calculated by finding the average difference between opponent's typical points scored vs what they scored against a given team. NOT a great example of a way to weigh defense, but an example of how more complex ideas can be implemented using more complicated programming practices.

```python
def defense(match,color,team):
    enemyColor = "B"
    if(color == "B"):
        enemyColor = "R"
    else:
        enemyColor = "B"
    returnval = 0
    counter = 0
    for index,value in df[(df.get("Match")==match) & (df.get("Color")==enemyColor)].get("points_earned").items():
        meanval = df.groupby("Team").mean().loc[index].get("points_earned")
        if(abs(value-meanval)<5):
            returnval += (value-meanval)
            counter+=1
        else:
            returnval += 0
    if(counter!=0):
        return returnval/counter
    else:
        return 0

df = df.reset_index().assign(defscore = df.get(["Match","points_earned","Color"]).reset_index().apply(lambda row: defense(row['Match'],row['Color'],row['Team']),axis=1))
```

```python
df.groupby("Team").mean().get("defscore").sort_values(ascending = False)
```
